## Import python modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy import stats 
import os
import sys
sys.path.append('../../lib/')
import constants as c


BOOL_SAVE_FILE = 0
```

## The protein length distributions for the different species

### Retrieving the statistical descriptions of the length's distributions

```python
# statistics on length distribution for different species
#########################################################
#c.G_NCBI_GENOME_DATA_FILE #c.WRONG_ANNOTATIONS_MERGED_FILE #c.STAT_MERGED_FILE
###############################################################################
#
stat_file = c.STAT_MERGED_FILE #c.G_NCBI_GENOME_DATA_FILE #c.WRONG_ANNOTATIONS_MERGED_FILE #c.STAT_MERGED_FILE
out_file = stat_file
print("The statistical descriptions of the protein distributions for the different species is in:\n", stat_file, "\n")

# retrieve data and diminish the number of columns
stat_df = pd.read_csv(stat_file, low_memory=False, sep="\t")

# visualize data
if 1:
    pd.set_option('display.max_columns', None)
    display(stat_df.head(2))
    print(stat_df.shape)
    print(stat_df.columns)
    if (stat_file == c.WRONG_ANNOTATIONS_MERGED_FILE) or (stat_file == c.STAT_MERGED_FILE):
        print(stat_df["merged_division_superregnum"].value_counts(), "\n")
    elif stat_file == c.G_NCBI_GENOME_DATA_FILE:  
        print(stat_df["division_8"].value_counts(), "\n")
    else:
        sys.exit(0)
```

## Modify the data frame

```python
df = stat_df.copy()
if (stat_file == c.WRONG_ANNOTATIONS_MERGED_FILE) or (stat_file == c.STAT_MERGED_FILE):
    df.loc[df.genes_division == "metazoa", 'genes_division'] = "invertebrates"
    df.loc[df.merged_division_superregnum == "metazoa", 'merged_division_superregnum'] = "invertebrates"
    df.loc[df.genes_division == "protist", 'genes_division'] = "protists"
    df.loc[df.merged_division_superregnum == "protist", 'merged_division_superregnum'] = "protists"
elif stat_file == c.G_NCBI_GENOME_DATA_FILE:    
    df.loc[df.division_7 == "metazoa", 'division_7'] = "invertebrates"
    df.loc[df.division_8 == "metazoa", 'division_8'] = "invertebrates"
else:
    sys.exit(0)
```

```python
if 1:
    pd.set_option('display.max_columns', None)
    display(stat_df.head(2))
    print(df.shape)
    print(df.columns)
    if (stat_file == c.WRONG_ANNOTATIONS_MERGED_FILE) or (stat_file == c.STAT_MERGED_FILE):
        print(df["merged_division_superregnum"].value_counts(), "\n")
    elif stat_file == c.G_NCBI_GENOME_DATA_FILE:    
        print(df["division_8"].value_counts(), "\n")
    else:
        sys.exit(0)
```
