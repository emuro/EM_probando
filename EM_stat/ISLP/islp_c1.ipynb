# Wage

```python
import matplotlib.pyplot as plt
import seaborn as sns

from ISLP import load_data
Wage = load_data('Wage')
if 1:
    display(Wage.shape, Wage.head(2), Wage.describe()) 
```

## Wage vs. Age

```python
sns.lmplot(x="age", y="wage", data=Wage, order=3, aspect=1., scatter_kws={"color": "blue", "alpha": 0.5, "s": 2.}, line_kws={"color": "red"})
plt.title("Wage vs. Age")
plt.xlabel("Age")
plt.ylabel("Wage")
plt.show()
```

## Wage vs. Year

```python
sns.lmplot(x="year", y="wage", data=Wage, order=1, aspect=1., scatter_kws={"color": "blue", "alpha": 0.5, "s": 2.}, line_kws={"color": "red"})
plt.title("Wage vs. Year")
plt.xlabel("Year")
plt.ylabel("Wage")
plt.show()
```

## Wage vs. Education (boxplot)

```python
plt.style.use('ggplot') # 'default'
Wage.boxplot("wage", by="education",
           notch=True,  # notch shape
           vert=True,  # vertical box alignment
           patch_artist=True, # fill with color
           color="blue" #change to different colors
          )
plt.xticks(rotation=30)
plt.title(None)
plt.xlabel("Education Level")
plt.ylabel("Wage")
plt.show()
```
