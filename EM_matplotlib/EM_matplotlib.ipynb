## Import some libs

```python
import numpy as np
from matplotlib import pyplot as plt
```

---

## Simple start 

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y, color='salmon')
#fig.save("sinusoidal.pdf")
plt.show()
```

---

## Simple pie chart

```python
x = [10, 30 , 40, 10]
labels = ["A", "T", "C", "G"]
fig, ax = plt.subplots()
plt.pie(x, labels=labels, labeldistance=0.5)
plt.show()
```

---

# Simple bar chart

```python
x = [20, 15, 5, 15, 30, 20]
y = [17, 10, 12, 10, 35, 25]
z = [15, 12, 18, 5, 27, 23]
labels = ["D", "E", "W", "T", "R", "G"]
w, c = 0.2, np.arange(len(labels))
                      
fig, ax = plt.subplots()
plt.bar(c-w, height=x, color='r', align='center', width=0.2, label='set1')
plt.bar(c,   height=y, color='b', align='center', width=0.2, label='set2')
plt.bar(c+w, height=z, color='g', align='center', width=0.2, label='set3')
ax.set_xticks(c)
ax.set_xticklabels(labels)
ax.set_ylabel("Percent")
ax.set_xlabel("Amino acid")
ax.set_title("Human")
ax.legend()
plt.show()
```
